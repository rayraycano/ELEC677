```python
def plot_results(RUN):
    train_file = open(RUN + '_train.txt', 'r')
    index_file = open(RUN + '_index.txt', 'r')
    val_file = open(RUN + "_test.txt", 'r')
    def f2i(intfile):
        return [float(x) for x in intfile.read().split(',') if x]
    x = np.array([int(x) for x in f2i(index_file)])
    train = np.array(f2i(train_file))
    val = np.array(f2i(val_file))
    plt.plot(x, train, '-r')
    plt.plot(x, val, '-g')
#     plt.plot(f2i(index_file), f2i(val_file))
    plt.show()
    train_file.close()
    index_file.close()
    val_file.close()
```
